```python
# Activation Function
  (Signal ko bhejna hai ya nahi)
  - make data non-linear (nonlinearity)
  - decision boundaries
  'brain-like' behaviour

Commonly used =>
1. ReLU (Rectified Linear Unit) =>
  - if input is positive keep it if input is negative make it zero.
  - flow the gradient.
  - fast and good in deep networks and in hidden layers.
  - Problem: Some neurons get stuck at 0 forever.
2. Sigmoid (Output is always between 0 and 1 like the Probability)
  - used for binary classification output layer. (binary mein output chahiye use Sigmoid)
  - anything you put in becomes probability.
  - verybad in hidden layers, beacause it vanishes the gradient.
3. Tanh (Hyperbolic tangent)
  - Output is between -1 and 1.(centered around 0 => benefit: network learns fast)
  - verybad in hidden layers, beacause it vanishes the gradient.
4. Softmax
  - used for multiclass outputs
  - Takes a list of numbers -> converts into a list of probabilities.
  - Total output of probabilities = 1.0
  - Example: [2.1 , 1.3 , 0.4] => 3.8
      2.1 / 3.8 , 1.3 / 3.8 , 0.4 / 3.8 => [0.552 , 0.342 , 0.105] => 0.998
  - Used in Output layer of multiclass classification
  - Used in NLP classification
  - Used in Image classification(dataset like CIFAR-10)
5. Leaky ReLU.. (ELU , GELU, ...)
    - relu problem: if neuron gets negative values forever, it dies.
    - leaky relu fixes this problem:
      if x < 0 => it gives a smallest negative slope (0.01x) (negligible value)
 and more.. (explore)
```

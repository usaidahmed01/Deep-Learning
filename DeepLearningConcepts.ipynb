```python
 Artificial Neuron / Perceptron:
  1. Takes input
  2. Combines them (with importance / weights)
  3. Decides the Output

Neuron:
  input
    x1 = age
    x2 = salary
    x3 = education level
    Weights w1 , w2 , w3

  Z = (x1 * w1) + (x2 * w2) + (x3 * w3) + b
  # b = internal biased (Internal Preference) artificial neurons have the internal bias

  Output = activation(Z)

  Activation funtion decides: should the neuron fire strong / weak / none
    1. Sigmoid
    2. ReLU (only positive values)

Inputs => Weighted Sum => Activation => Output
Output = f(w.x + b)
Where f is the Activation Function

# 3-Tiers of Layer in Neural Networks
  1. Input Layer
    - Just Data Coming in
    - No Calculation Here just passing values
  2. Hidden Layers
    - This is where the learning happens
    - Each Neuron here:
      . Receives Input
      . Applies weight + bias
      . Passes outputs through activation()
    - Hidden layer are called hidden because they are not directly visible      (but they learn patterns)
  3. Output Layer
    - This Gives us:
      . A probabilty of classification (employee attrition)
      . Or a value for regression
```

```python
 Input Processing Output -> called forward Propagation

Perceptron makes the static data into dynamic
```

```python
# Perceptron Model
import numpy as np
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score, classification_report
```

```python
data = load_diabetes()
X = data.data
y = (data.target > data.target.mean()).astype(int) # Convert into 0 and 1
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5 , random_state= 42)

model = Perceptron() # Max heights , penality , alpha
model.fit(X_train, y_train)
```

```python
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

print('Training Accuracy: ', accuracy_score(y_train, y_pred_train))
print('Training Accuracy: ', accuracy_score(y_test, y_pred_test))
```

```python
# 11-NOV-2025

from keras.datasets import mnist
(train_images , train_labels) , (test_images , test_labels) = mnist.load_data()
```

```python
# Network Architecture

from tensorflow import keras
from keras import layers
model = keras.Sequential([
    layers.Dense(512 , activation = 'relu'), # 512 Neurons , 784(28 x 28) features in our case
    layers.Dense(10 , activation = 'softmax') # possible outputs 10
]) # Sequential Model (input -> hidden -> output)
# Dense means: One feature is fully connected with every neuron and each neuron in hidden layer is fully connected with another neuron
# Activation Functions used to convert the discrete signals into continuous
```

```python
model.compile(optimizer="rmsprop", # har epoch pr weights readjust karta hai or dekhta hai ke comparison karke ke output ke kitna near hai
              loss = "sparse_categorical_crossentropy",
              metrics = ["accuracy"])
```

```python
train_images = train_images.reshape((60000 , 28 * 28)) # flatten the array
train_images = train_images.astype("float32") / 255

test_images = test_images.reshape((10000 , 28 * 28))
test_images = test_images.astype("float32") / 255
```

```python
model.fit(train_images , train_labels , epochs = 5 , batch_size = 128) # epochs => iteration , RAM batch size = 128 ek saath pooray 60000 nahi desakti RAM
```

```python
test_digits = test_images[0:10]
predictions = model.predict(test_digits)
```

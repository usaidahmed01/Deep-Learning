```python
import numpy as np
from keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense, GRU

# For texts we use Sequential
```

```python
vocab_size = 10000
(x_train , y_train) , (x_test , y_test) = imdb.load_data(num_words=vocab_size)
```

```python
max_length = 200
X_train = pad_sequences(x_train, max_length)
X_test = pad_sequences(x_test, max_length)
```

```python
model_LSTM = Sequential([
    Embedding(input_dim = vocab_size, output_dim = 128 , input_length = max_length),
    LSTM(128), # Memory Cells
    Dense(1, activation='sigmoid')
])
```

```python
model_LSTM.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
model_LSTM.fit(X_train , y_train , epochs = 3 , batch_size = 64 , validation_split = 0.2)
# ANOTHER WAY TO CHECK THE VALIDATION SPLIT
```

```python
model_GRU = Sequential([
    Embedding(input_dim = vocab_size, output_dim = 128 , input_length = max_length),
    GRU(128), # Memory Cells
    Dense(1, activation='sigmoid')
])

model_GRU.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)

model_GRU.fit(X_train , y_train , epochs =3 , batch_size = 64 , validation_split = 0.2)
```

```python
LSTM_loss , LSTM_acc = model_LSTM.evaluate(X_test , y_test)
GRU_loss , GRU_acc = model_GRU.evaluate(X_test , y_test)
```
